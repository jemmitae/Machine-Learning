This particular Automobile Data Set includes a good mix of categorical values as well as continuous values and serves as a useful example that is relatively easy to understand. Since domain understanding is an important aspect when deciding how to encode various categorical values - this data set makes a good case study.

Read the data into Pandas

```python
import pandas as pd

# Define the headers since the data does not have any
headers = ["symboling", "normalized_losses", "make", "fuel_type", "aspiration",
           "num_doors", "body_style", "drive_wheels", "engine_location",
           "wheel_base", "length", "width", "height", "curb_weight",
           "engine_type", "num_cylinders", "engine_size", "fuel_system",
           "bore", "stroke", "compression_ratio", "horsepower", "peak_rpm",
           "city_mpg", "highway_mpg", "price"]

# Read in the CSV file and convert "?" to NaN
df = pd.read_csv("http://mlr.cs.umass.edu/ml/machine-learning-databases/autos/imports-85.data",
                  header=None, names=headers, na_values="?" )
df.head()
```

Se toman los missing y se arreglan con la media de los datos, en este caso el campo normalized_losses esto debido a que se evidencia que si se elimina este campo se estaria quitando el 20% de los datos.

```python
df.normalized_losses.fillna(df.normalized_losses.median(), inplace=True)
```

```python
df.stroke.fillna(df.stroke.median(), inplace=True)
```

```python
df.price.fillna(df.price.median(), inplace=True)
```

# Exercise 13.2

Split the data into training and testing sets

Train a Random Forest Regressor to predict the price of a car not using the nominal features

```python
Num_df = df.select_dtypes(exclude=['object']).copy()
Num_df.head()
```

Se generan X y y, teniendo en cuenta unicamente los datos numericos.

```python
X = Num_df.drop('price', axis=1)
y = Num_df.price
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=1)
```

```python
from sklearn.ensemble import RandomForestRegressor
regr = RandomForestRegressor(max_depth=2, random_state=0, n_estimators=100)
```

```python
regr.fit(X, y)
```

```python
pd.DataFrame({'feature':X, 'importance':regr.feature_importances_}).sort_values('importance')
```

```python
predicted = regr.predict(X_test)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted))
print('RMSE:')
print(rmse)
```

# Exercise 13.4

Apply two other methods of categorical encoding

compare the results

```python
X_a = ce.HelmertEncoder().fit_transform(df.drop(['price'], axis=1))
X_b = ce.HashingEncoder(n_components=8).fit_transform(df.drop(['price'], axis=1))
```

```python
regr.fit(X_a, y)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_a, y, test_size=0.30, random_state=1)
predicted = regr.predict(X_test)
import numpy as np
from sklearn.metrics import mean_squared_error
rmse_a = np.sqrt(mean_squared_error(y_test, predicted))
print('RMSE_a Encode Helmert:')
print(rmse_a)
```

```python
regr.fit(X_b, y)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_b, y, test_size=0.30, random_state=1)
predicted = regr.predict(X_test)
import numpy as np
from sklearn.metrics import mean_squared_error
rmse_b = np.sqrt(mean_squared_error(y_test, predicted))
print('RMSE_b Encode Hashing:')
print(rmse_b)
```

```python
print('RMSE Numericas:',rmse,'RMSE Binary:',rmse2,'RMSE Helmert:',rmse_a,'RMSE Hashing:', rmse_b)
```
